```python
import pandas as pd
```

```python
global_temps_df = pd.read_csv('global_temps.csv') # 연도에 따른 월별 지구의 평균 표면 온도 데이터셋
global_temps_df
```

```python
# 월 라벨들을 아래로 긴 형식으로 변환
global_temps_df = global_temps_df.melt(id_vars=['Year', 'J-D', 'D-N', 'DJF', 'MAM', 'JJA', 'SON'], var_name='Month', value_name='Temp')

# 월 라벨들을 각각 숫자로 맵핑한 딕셔너리 생성
months_dict = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
    'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}

# 맵핑 적용
global_temps_df['Month'] = global_temps_df['Month'].map(months_dict)

global_temps_df
```

```python
ozoneHole_df = pd.read_csv('OzoneHole_Data.csv') # 연도에 따른 오존층과 오존 구멍의 면적 데이터셋
ozoneHole_df
```

```python
seaice_df = pd.read_csv('seaice.csv') # 연도에 따른 빙하 면적 데이터셋
seaice_df.columns = seaice_df.columns.str.strip() # 열 라벨 공백 제거
seaice_df
```

```python
# 표면 온도, 오존 프레임 병합
merged_df = pd.merge(global_temps_df, ozoneHole_df, on=['Year'], how='outer')
merged_df
```

```python
# 빙하 면적 프레임 병합
merged_df = pd.merge(merged_df, seaice_df, on=['Year', 'Month'], how='outer')

# 열 라벨 순서 재조정
merged_df = merged_df.reindex(columns=['Year', 'Month', 'Day', 'Temp', 'J-D', 'D-N', 'DJF', 'MAM', 'JJA', 'SON',
                                       'Hole Area', 'Minimum Ozone', 'Extent', 'Missing', 'Source Data', 'hemisphere'])

merged_df
```

```python
aggi_df = pd.read_csv('AGGI_Table.csv') # 연도에 따른 대기 중 온실가스 농도의 지수 데이터셋

# 열 라벨 이름 수정
aggi_df = aggi_df.rename(columns={'Total.1': 'CO2-eq', 'CFC*': 'CFC', 'HFCs*': 'HFCs', 'change **': 'change'})

aggi_df
```

```python
# 온실가스 농도 프레임 병합
merged_df = pd.merge(merged_df, aggi_df, on=['Year'], how='outer')

merged_df.to_csv('merged_df.csv', index=False) # 모든 데이터셋을 병합한 프레임을 파일로 저장

merged_df
```
